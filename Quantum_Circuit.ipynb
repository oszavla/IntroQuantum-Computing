```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.circuit.library import UGate
from numpy import pi
from qiskit.visualization import plot_histogram
from qiskit.result import marginal_distribution
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit import transpile
from qiskit_ibm_runtime import SamplerV2 as Sampler
import numpy as np

# Define the statevector
statevector = np.array([1/2, np.sqrt(3)/2])  # [1/2, sqrt(3)/2]

# Create a quantum circuit with 1 qubit
qc = QuantumCircuit(1)

# Initialize the qubit in the state |ϕₐ⟩
qc.initialize(statevector, 0)

# Measure the qubit (optional)
qc.measure_all()

# Simulate the circuit

result = AerSimulator(shots=1e7).run(qc).result()
counts = result.get_counts(qc)

print(counts)
display(plot_histogram(counts))
qc.draw('mpl')
```
